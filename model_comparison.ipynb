```python
import matplotlib.pyplot as plt
import numpy as np

# ===============================
# Data (in % share or normalized)
# ===============================

# Model 1 (percent shares directly)
model1 = {
    "Gas": 28.4,
    "Diesel": 14.9,
    "Wind Onshore": 28.4,
    "PV": 28.4
}

# Model 2 (MW values → convert to shares)
# Each technology has 381.4 MW
model2_raw = {
    "Gas": 381.4,
    "Diesel": 381.4,
    "Wind Onshore": 381.4,
    "PV": 381.4
}

total_m2 = sum(model2_raw.values())
model2 = {k: v / total_m2 * 100 for k, v in model2_raw.items()}

# Model 3 (percent shares)
model3 = {
    "PV": 58.8,
    "Wind Onshore": 14.0,
    "Gas": 27.2
    # Diesel omitted → 0%
}

# Ensure all technologies appear for all models
techs = ["Gas", "Diesel", "Wind Onshore", "PV"]

def get_vals(model_dict, techs):
    return [model_dict.get(t, 0) for t in techs]

vals_m1 = get_vals(model1, techs)
vals_m2 = get_vals(model2, techs)
vals_m3 = get_vals(model3, techs)

# ===============================
# Plot
# ===============================

bar_width = 0.25
x = np.arange(len(techs))

plt.figure(figsize=(8, 6))

plt.bar(x - bar_width, vals_m1, width=bar_width, label="Model 1", color="#4C72B0")
plt.bar(x, vals_m2, width=bar_width, label="Model 2", color="#55A868")
plt.bar(x + bar_width, vals_m3, width=bar_width, label="Model 3", color="#C44E52")

# Axis settings
plt.xticks(x, techs, fontsize=13)
#plt.ylabel("", fontsize=14, rotation=90)
plt.xlabel("Technology", fontsize=14)
plt.title("Share of Energy Mix (%)", fontsize=14, loc="left")

plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ===============================
# Data (in % share or normalized)
# ===============================

# Model 1 (percent shares directly)
model1 = {
    "Gas": 28.4,
    "Diesel": 14.9,
    "Wind Onshore": 28.4,
    "PV": 28.4
}

# Model 2 (MW values → convert to %)
model2_raw = {
    "Gas": 381.4,
    "Diesel": 381.4,
    "Wind Onshore": 381.4,
    "PV": 381.4
}
total_m2 = sum(model2_raw.values())
model2 = {k: v / total_m2 * 100 for k, v in model2_raw.items()}

# Model 3 (percent shares)
model3 = {
    "PV": 58.8,
    "Wind Onshore": 14.0,
    "Gas": 27.2
}

# Technologies in consistent order
techs = ["Gas", "Diesel", "Wind Onshore", "PV"]

def get_vals(model, techs):
    return [model.get(t, 0) for t in techs]

vals_m1 = get_vals(model1, techs)
vals_m2 = get_vals(model2, techs)
vals_m3 = get_vals(model3, techs)

# ===============================
# Plot setup
# ===============================

bar_width = 0.23
x = np.arange(len(techs))

plt.figure(figsize=(13, 6))

bars1 = plt.bar(x - bar_width, vals_m1, width=bar_width, label="Model 1", color="#4C72B0")
bars2 = plt.bar(x, vals_m2, width=bar_width, label="Model 2", color="#55A868")
bars3 = plt.bar(x + bar_width, vals_m3, width=bar_width, label="Model 3", color="#C44E52")

# ===============================
# Add labels inside bars
# ===============================

def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            plt.text(
                bar.get_x() + bar.get_width()/2,
                height * 0.5,                     # Middle of the bar
                f"{height:.1f}%",
                ha="center", va="center",
                fontsize=11, color="white", fontweight="bold"
            )

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# ===============================
# Axes & styling
# ===============================

plt.xticks(x, techs, fontsize=13)
plt.ylabel("Share of Energy Mix (%)", fontsize=14)
plt.xlabel("Technology", fontsize=14)
plt.title("Final Energy Mix Across Models", fontsize=16, fontweight='bold')

plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ===============================
# Data
# ===============================

model1 = {
    "Gas": 28.4,
    "Diesel": 14.9,
    "Wind Onshore": 28.4,
    "PV": 28.4
}

model2_raw = {
    "Gas": 381.4,
    "Diesel": 381.4,
    "Wind Onshore": 381.4,
    "PV": 381.4
}
total_m2 = sum(model2_raw.values())
model2 = {k: v / total_m2 * 100 for k, v in model2_raw.items()}

model3 = {
    "PV": 58.8,
    "Wind Onshore": 14.0,
    "Gas": 27.2
}

techs = ["Gas", "Diesel", "Wind Onshore", "PV"]

def get_vals(model, techs):
    return [model.get(t, 0) for t in techs]

vals_m1 = get_vals(model1, techs)
vals_m2 = get_vals(model2, techs)
vals_m3 = get_vals(model3, techs)

# ===============================
# Plot
# ===============================

fig, ax = plt.subplots(figsize=(9,6))
bar_width = 0.25
x = np.arange(len(techs))

bars1 = ax.bar(x - bar_width, vals_m1, width=bar_width, label="Model 1", color="#4C72B0")
bars2 = ax.bar(x, vals_m2, width=bar_width, label="Model 2", color="#55A868")
bars3 = ax.bar(x + bar_width, vals_m3, width=bar_width, label="Model 3", color="#C44E52")

# ===============================
# Add minimalistic labels
# ===============================

ax.bar_label(bars1, fmt="%.1f%%", padding=2, fontsize=11)
ax.bar_label(bars2, fmt="%.1f%%", padding=2, fontsize=11)
ax.bar_label(bars3, fmt="%.1f%%", padding=2, fontsize=11)

# ===============================
# Styling
# ===============================

ax.set_xticks(x)
ax.set_xticklabels(techs, fontsize=14)
ax.set_xlabel("Technology", fontsize=14)
ax.set_title("Share of Energy Mix (%)", fontsize=14, loc="left")

ax.grid(axis="y", linestyle="--", alpha=0.5)
ax.legend(fontsize=14)
## Increase Y-tick fontsize
ax.tick_params(axis='y', labelsize=14)

plt.tight_layout()
plt.show()
```
